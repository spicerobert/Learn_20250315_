- 建立正弦波


```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸數據，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 計算正弦波的 y 軸數據
y_sin = np.sin(x)

# 計算餘弦波的 y 軸數據
y_cos = np.cos(x)

# 使用 matplotlib 繪製圖形
plt.plot(x, y_sin, label='sin(x)')  # 繪製正弦波，並設定標籤
plt.plot(x, y_cos, label='cos(x)')  # 繪製餘弦波，並設定標籤

# 設定圖形標題和軸標籤
plt.title('正弦波和餘弦波圖形')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖例
plt.legend()

# 顯示圖形
plt.show()
```
