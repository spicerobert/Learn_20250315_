- 建立正弦波


```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸數據，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 計算 y 軸數據，即 sin(x)
y = np.sin(x)

# 使用 matplotlib 繪製圖形
plt.plot(x, y)

# 設定圖形標題和軸標籤
plt.title('正弦波圖形')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 顯示圖形
plt.show()
```
