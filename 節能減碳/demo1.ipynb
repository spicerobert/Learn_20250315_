- 建立正弦波


```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸數據，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 計算 y 軸數據，即 sin(x) 和 cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 使用 matplotlib 繪製圖形
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')

# 設定圖形標題和軸標籤
plt.title('sin & cos')
plt.xlabel('x')
plt.ylabel('值')

# 顯示圖例
plt.legend()

# 顯示圖形
plt.show()
```
